```python
xraw = np.random.random(1000)
```

```python
plt.plot(xraw,'.')
```

```python
time = range(0,len(xraw))
plt.plot(time, xraw,'.')
```

```python
price = xraw.cumsum()
```

```python
plt.plot(time,price)
```

```python
clean_price = (xraw - 0.5).cumsum()
```

```python
plt.plot(time,clean_price)
```

```python
clean_prices_2 = (xraw - xraw.mean()).cumsum()
```

```python
plt.plot(time,clean_prices_2)
```

```python
xraw.mean()
```

```python
#seed allows us to always use the same random numbers in a cell.
np.random.seed(22)
plt.plot(time,(np.random.random(1000) - 0.5).cumsum())
```
